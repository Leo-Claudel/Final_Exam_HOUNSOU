```python
# Step 3: Data Cleaning: Here we will clean the data by converting relevant columns to appropriate data types and handling any inconsistencies.
# Remove commas from numeric columns and convert to float
df['Market value as of 31 December, 2023'] = df['Market value as of 31 December, 2023'].replace(',', '', regex=True).astype(float)
df['Number of shares'] = df['Number of shares'].replace(',', '', regex=True).astype(float)


```

```python
# Display cleaned data types
print("Cleaned data types:\n", df.dtypes)
```

```python
# Step 4: Exploratory Data Analysis (EDA): We will analyze the data and extract insights.

# Get top 10 companies by market value
top_companies = df.nlargest(10, 'Market value as of 31 December, 2023')
```

```python
top_companies
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='Market value as of 31 December, 2023', y='Company', data=top_companies, palette='viridis')
plt.title('Top 10 Companies by Market Value')
plt.xlabel('Market Value (in billions)')
plt.ylabel('Company')
plt.show() 

# Top Companies: This bar chart shows the top 10 companies by market value, highlighting which ones dominate the market.
```

```python
# Insight 2: Portfolio Distribution
# Visualization of portfolio distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['% of total portfolio'], bins=20, kde=True, color='blue')
plt.title('Distribution of % of Total Portfolio')
plt.xlabel('% of Total Portfolio')
plt.ylabel('Frequency')
plt.show()

# Portfolio Distribution: We have here an illustration of the distribution of the percentage of the total portfolio, indicating how the investments are diversified .
```

```python
# Insight 3: Market Value vs. Number of Shares
# Scatter plot of Market Value vs. Number of Shares
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number of shares', y='Market value as of 31 December, 2023', data=df, hue='Domain', palette='Set2')
plt.title('Market Value vs. Number of Shares')
plt.xlabel('Number of Shares')
plt.ylabel('Market Value')
plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')  # Log scale for better visualization
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Market Value vs. Shares: This scatter plot reveals the relationship between the number of shares and market value, categorized by domain.
```

```python
# Insight 4: Average Score by Domain
# Calculate average score by domain
avg_score_by_domain = df.groupby('Domain')['Score'].mean().reset_index()

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Domain', data=avg_score_by_domain, palette='coolwarm')
plt.title('Average Score by Domain')
plt.xlabel('Average Score')
plt.ylabel('Domain')
plt.show()

# Average Score by Domain: This bar chart provides insight into the average score by domain, indicating the performance across different sectors.
```

```python
# Insight 5: Correlation Heatmap
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df[['Market value as of 31 December, 2023', 'Number of shares', '% of total portfolio', 'Score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Correlation Analysis: This heatmap shows the correlations between key numerical variables, helping to identify relationships within the data.
```
